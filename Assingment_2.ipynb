```python

from google.colab import files
uploaded = files.upload()

import io

import pandas as pd
data = pd.read_csv("Pokemon.csv")
data.head(n=10)
          

df2 = pd.read_csv(io.BytesIO(uploaded['Pokemon.csv']))


speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "speed Low"
    elif val == speed_mean:
        return "speed neutral"
    else:
        return "speed High"


data["speed_high_low"] = data["Speed"].apply(set_speed)        
```

```python
hp_mean = data["HP"].mean()

def set_hp(val):
    if val < hp_mean:
        return "hp Low"
    elif val == hp_mean:
        return "hp neutral"
    else:
        return "hp High"


data["hp_high_low"] = data["HP"].apply(set_hp)        
```
